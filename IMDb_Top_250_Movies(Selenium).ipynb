```python
#LINK_ALL


result_all = driver.find_element_by_css_selector('#main > div > span > div > div')
result_all

html_all = result_all.get_attribute('innerHTML')
print(html_all)
```

```python
link = result_all.find_element_by_class_name('titleColumn').find_element_by_tag_name('a').get_attribute('href')
print(link)
```

```python
list_ = result_all.find_element_by_class_name('lister-list').find_elements_by_tag_name('tr')
ids = []
imdbLinks = []
smallPoster = []
for movie in list_:
    link = movie.find_element_by_class_name('titleColumn').find_element_by_tag_name('a').get_attribute('href')
    posterLink = movie.find_element_by_class_name('posterColumn').find_element_by_tag_name('a').find_element_by_tag_name('img').get_attribute('src')
    imdbLinks.append(link)
    smallPoster.append(posterLink)
    id_ = link.split('/')[4]
    ids.append(id_)

notseen = []
seen = []

print('Done!')
```
